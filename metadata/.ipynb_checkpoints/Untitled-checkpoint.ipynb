```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import mplfinance as mpf

# Replace 'HDFCBANK.NS' with the desired stock symbol
symbol = 'HDFCBANK.NS'

# Download historical data using yfinance
data = yf.download(symbol, start='2023-01-01', end='2023-12-29')

# Calculate moving averages, momentum, and RSI
data['SMA60'] = data['Close'].rolling(window=60).mean()
data['SMA80'] = data['Close'].rolling(window=80).mean()
data['Momentum'] = data['Close'].pct_change() * 100

# Plotting
fig, axlist = mpf.plot(data, type='candle', mav=(60, 80),
                       volume=True, show_nontrading=True,
                       addplot=[
                           mpf.make_addplot(data['Momentum'], panel=1, color='b', secondary_y=False)
                       ],
                       title=f"Visualizing a strategy on {symbol}",
                       ylabel='Price',
                       ylabel_lower='Volume & Indicators',
                       figratio=(10, 6),
                       )
plt.show()
```

```python
import plotly.graph_objects as go
import yfinance as yf
from plotly.subplots import make_subplots

# Replace 'HDFCBANK.NS' with the desired stock symbol
symbol = 'HDFCBANK.NS'

# Download historical data using yfinance
hist = yf.download(symbol, start='2023-07-01', end='2023-12-29')
hist['diff'] = hist['Close'] - hist['Open']
hist.loc[hist['diff']>=0, 'color'] = 'green'
hist.loc[hist['diff']<0, 'color'] = 'red'


fig3 = make_subplots(specs=[[{"secondary_y": True}]])
fig3.add_trace(go.Candlestick(x=hist.index,
                              open=hist['Open'],
                              high=hist['High'],
                              low=hist['Low'],
                              close=hist['Close'],
                              name='Price'))
fig3.add_trace(go.Scatter(x=hist.index,y=hist['Close'].rolling(window=20).mean(),marker_color='blue',name='20 Day MA'))
fig3.add_trace(go.Scatter(x=hist.index,y=hist['Close'].rolling(window=40).mean(),marker_color='yellow',name='40 Day MA'))
fig3.add_annotation(
    dict(
        font=dict(color='black', size=15, family='Arial'),
        x=0.5,  # Center aligned horizontally
        y=-0.12,
        showarrow=False,
        text="A very clear explanation",
        textangle=0,
        xanchor='center',  # Center aligned horizontally
        yanchor='bottom',  # Aligned to the bottom
        xref="paper",
        yref="paper"
    )
)



fig3.add_trace(go.Bar(x=hist.index, y=hist['Volume'], name='Volume', marker={'color':hist['color']}),secondary_y=True)
fig3.update_yaxes(range=[0,hist['Volume'].max() * 5],secondary_y=True)
fig3.update_yaxes(visible=False, secondary_y=True)
fig3.update_layout(xaxis_rangeslider_visible=False)  #hide range slider
fig3.update_layout(title={'text':symbol, 'x':0.5})

fig3.update_xaxes(rangebreaks = [ dict(bounds=['sat','mon']) ] ) # hide weekends 
fig3.show()
```

```python
symbols = ['HDFCBANK.NS', 'INFY.NS', 'TCS.NS', 'LICI.NS', 'BHARTIARTL.NS']

for symbol in symbols:
    hist1 = yf.download(symbol, start='2023-01-01', end='2023-12-29')
    
    hist1['20DMA']=hist1['Close'].rolling(window=20).mean()
    hist1['40DMA']=hist1['Close'].rolling(window=40).mean()
    hist1['100DMA']=hist1['Close'].rolling(window=100).mean()
    hist1['diff'] = hist1['Close'] - hist1['Open']
    hist1.loc[hist1['diff']>=0, 'color'] = 'green'
    hist1.loc[hist1['diff']<0, 'color'] = 'red'
    
    hist = hist1[hist1.index > (pd.to_datetime('today') - pd.DateOffset(days=180))]
    
    

    fig3 = make_subplots(specs=[[{"secondary_y": True}]])
    fig3.add_trace(go.Candlestick(x=hist.index,
                              open=hist['Open'],
                              high=hist['High'],
                              low=hist['Low'],
                              close=hist['Close'],
                              name='Price'))
    fig3.add_trace(go.Scatter(x=hist.index,y=hist['20DMA'],marker_color='blue',name='20 Day MA'))
    fig3.add_trace(go.Scatter(x=hist.index,y=hist['40DMA'],marker_color='orange',name='40 Day MA'))
    fig3.add_trace(go.Scatter(x=hist.index,y=hist['100DMA'],marker_color='green',name='100 Day MA'))
    
    fig3.add_trace(go.Bar(x=hist.index, y=hist['Volume'], name='Volume', marker={'color':hist['color']}),secondary_y=True)
    fig3.update_yaxes(range=[0,hist['Volume'].max() * 5],secondary_y=True)
    fig3.update_yaxes(visible=False, secondary_y=True)
    fig3.update_layout(xaxis_rangeslider_visible=False)  #hide range slider
    fig3.update_layout(title={'text':symbol, 'x':0.5})

    fig3.update_xaxes(rangebreaks = [ dict(bounds=['sat','mon']) ] ) # hide weekends 
    fig3.show()
```
